```python
#import required librariers
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import os

%matplotlib inline
```

```python
# Insert the path to image
image_path = '/content/Historical_places.jpeg'

# Read the image
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert from cv's BGR default color order to RGB

orb = cv2.ORB_create()  # Create ORB object
key_points, description = orb.detectAndCompute(img, None)  # Detect keypoints and compute descriptors
img_keypoints = cv2.drawKeypoints(img, key_points, img, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)  # Draw keypoints
plt.figure(figsize=(16, 16))
plt.title('ORB Interest Points')
plt.imshow(img_keypoints)
plt.axis('off')  # Hide axis
plt.show()
```
